```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split

# Verilerin saklandığı ana klasör
data_dir = 'data'
img_size = 28

# Veri ve etiket listeleri
data = []
labels = []

# Klasörleri dolaş
for label in range(10):
    path = os.path.join(data_dir, str(label))
    for img_name in os.listdir(path):
        # .ipynb_checkpoints klasörünü atla
        if img_name == '.ipynb_checkpoints':
            continue
        
        img_path = os.path.join(path, img_name)
        
        # Görüntüyü oku ve kontrol et
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Hata: Görüntü yüklenemedi - {img_path}")
            continue
        
        # Görüntüyü yeniden boyutlandır
        img = cv2.resize(img, (img_size, img_size))
        data.append(img)
        labels.append(label)

# Veriyi normalize et
data = np.array(data) / 255.0
data = data.reshape(-1, img_size, img_size, 1)
labels = np.array(labels)

# Veriyi kaydet
np.save('data.npy', data)
np.save('labels.npy', labels)

# Veri setini yükle
data = np.load('data.npy')
labels = np.load('labels.npy')

# Etiketleri kategorik hale getir (one-hot encoding)
labels = tf.keras.utils.to_categorical(labels, 10)

# Veriyi eğitim ve test setlerine ayır
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Veri artırma (data augmentation) tanımlaması
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=False,
    fill_mode='nearest'
)

datagen.fit(x_train)

# Modeli oluştur
model = models.Sequential([
    layers.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Modeli derle
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Modeli eğit
# ImageDataGenerator ile veri artırma kullanarak eğitim yapma
model.fit(datagen.flow(x_train, y_train, batch_size=32),
          epochs=20,
          validation_data=(x_test, y_test))

# Test seti üzerinde değerlendirme yap
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'Test doğruluğu: {test_acc}')

import matplotlib.pyplot as plt

# Test setinden bir görüntü seç
img = x_test[2]

# Görüntüyü tahmin et
predictions = model.predict(np.expand_dims(img, axis=0))
predicted_label = np.argmax(predictions)

# Sonuçları göster
plt.imshow(img.reshape(28, 28), cmap=plt.cm.binary)
plt.title(f'Tahmin Edilen: {predicted_label}')
plt.show()

# Birkaç görüntüyü göstermek için
num_images = 10  # Görüntü sayısını belirleyin
plt.figure(figsize=(10, 10))

for i in range(num_images):
    img = x_test[i]
    predictions = model.predict(np.expand_dims(img, axis=0))
    predicted_label = np.argmax(predictions)
    
    plt.subplot(1, num_images, i + 1)
    plt.imshow(img.reshape(28, 28), cmap=plt.cm.binary)
    plt.title(f'Tahmin: {predicted_label}')
    plt.axis('off')

plt.show()
```
